```python
import pandas as pd
```

```python
video = pd.read_csv('video_info_v2.csv')
```

```python
video
```

```python
stop_words = ["‚úÖ", "(", ')', '!', '-', '#', '–≤', '–∏', '–∏–ª–∏', '-', '.', ',', ':', '|', '/', 'üòç', 'üé∂', 'üéÅ', 'ü¶Å', 'üò±', 'üöó', 'üê∂', 'üòª', 'üòé', '‚ù§Ô∏è', 'üòÖ', 'üëç', '@', '‚ò£Ô∏è', 'üî•', '‚ö°', '–æ—Ç', '–Ω–∞', '–∫–∞–∫', '–∏–∑', '?', 'c', '–∫–∞–∫', '—á—Ç–æ', '–≤', '–∑–∞', '–Ω–µ', '"', "'", '*', 'üåõ', 'üò¥', 'üçå', 'üëê', 'üîî']
```

```python
%%time
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(stop_words=stop_words)
x = vectorizer.fit_transform(video['title'])
```

```python
%%time
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=50, random_state=42)  # –ó–∞–¥–∞–µ–º 2 –∫–ª–∞—Å—Ç–µ—Ä–∞
kmeans.fit(x)
```

```python
video['label'] = kmeans.labels_
video
```

```python
video.to_csv('video_info_v3.csv', index=False)
```
